### The formula that will be used to calculate complexity

O(V\*timeMod\*K):
 - V - інтенсивність подій
 - timeMod - час моделювання
 - K - кількість операцій для обробки однієї події

```python
def O(V, N, timeMod = 1, K = 1):
    return V * N * timeMod * K

V = 1

stats = dict()

for N in range(10, 301, 10):
    stats[N] = O(V, N)

import matplotlib.pyplot as plt

plt.plot(stats.keys(), stats.values())
plt.xlabel('N')
plt.ylabel('O')
plt.title('O(N)')
plt.show()
```
